```python
from bs4 import BeautifulSoup
import requests
from IPython.core.display import display, HTML 
# display(HTML(my_html)) 
import time
```

```python
source_code = requests.get('https://en.wikipedia.org/wiki/List_of_National_Basketball_Association_career_scoring_leaders')
```

```python
soup = BeautifulSoup(source_code.text, 'html5lib')
```

```python
print(soup.prettify)
```

```python
for i in soup.find_all(class_='wikitable sortable jquery-tablesorter'):
    print(i, '\n')
```
