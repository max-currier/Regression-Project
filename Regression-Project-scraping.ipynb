```python
from selenium import webdriver
chromedriver = "/Applications/chromedriver"
```

```python
driver = webdriver.Chrome(chromedriver)
```

```python
driver.get('https://www.nba.com/stats/alltime-leaders/')
```

```python
driver.page_source[:1000]
```

```python
stat_table = driver.find_element_by_xpath("//div[@class='nba-stat-table']")
```

```python
stat_table.text
```

```python
soup = BeautifulSoup(stat_table.text, 'lxml')
print(soup.prettify)
```

```python
df.to_csv('NBA Player Table')
```
